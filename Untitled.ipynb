Покдлючение библиотеки pandas

```python
import pandas as pd
```

Происходит загрузка датасета по варианту, который включает в себя 4344 строки и 8 столбцов, также выводим с помощью head() первые 20 строк датасета

```python
df = pd.read_csv('C:\Cads/1auto.csv')
df.head(20)
```

С помощью метода .info оцениваем данные, можно сделать вывод, что в стобце year находятся данные, которые не являются ни int, ни float,
все остальные типы данных соответствуют действительности и имеют тип object. 

```python
df.info()
```

С помощью .rename выявляем проблемы с названиями, если они есть и устраняем, делаем все названия с маленькой буквы, если название содержит 
несколько слов, разделяем их "_" и делаем оба слова с маленькой буквы

```python
df = df.rename(columns = {'Name': 'name', 'SellingPrice':'selling_price', 'seller_Type':'seller_type'})
df
```

Необходимо выявить пропуски и удалить их. в целом пропуски можно выявить с помощью проверки столбцов на типы, но мы для отработки данного навыка сделаем возможное удаление пропусков для каждого столбца.

```python
df['name'] = df['name'].fillna(0)
df['year'] = df['year'].fillna(0)
df['selling_price'] = df['selling_price'].fillna(0)
df['kmdriven'] = df['kmdriven'].fillna(0)
df['fuel'] = df['fuel'].fillna(0)
df['seller_type'] = df['seller_type'].fillna(0)
df['transmission'] = df['transmission'].fillna(0)
df['owner'] = df['owner'].fillna(0)
df
```

Необходимо выполнить проверку данных на наличие дубликатов, если они есть – удалить. Все это можно выполнить с помощью
drop_duplicates() и reset_index().

```python
df = df.drop_duplicates()
df = df.drop_duplicates().reset_index()
df
```

Тип данных не соответствует в столбце year, поэтому мы замяняем тип на необходимый и с помощью .info() проверяем есть 
ли еще какие-то не соответствия, после выполнения df.info() можно прийти к выводу, что теперь все типы 
столбцов соответвуют действительности

```python
df['year'] = pd.to_numeric(df['year'], errors = 'coerce')
df['year'] = 2007
df.info()
```

Необходимо сделать сводную таблицу и отсортировать ее по стобцу, чтоб пользователю было удобно определять, какая машина самая дешевая, какая самая дорогая и пользоваться таблицей так, как ему будет это удобно. Также было сделано 2 группировки с помощью group_by и
pivot_table.

```python
data_pivot = pd.pivot_table(df, index=['name'], values = ['selling_price'])
sorted_df = data_pivot.sort_values(by='selling_price')
sorted_df
```
