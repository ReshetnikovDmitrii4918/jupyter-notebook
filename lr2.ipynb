С помощью .rename выявляем проблемы с названиями, если они есть и устраняем, делаем все названия с маленькой буквы, если название содержит несколько слов, разделяем их "_" и делаем оба слова с маленькой буквы. Размер датасета составляет 925 строк на 12 колонок

Необходимо выявить пропуски и удалить их. в целом пропуски можно выявить с помощью проверки столбцов на типы, но мы для отработки данного навыка сделаем возможное удаление пропусков для каждого столбца.

Исследовать взаимосвязь между переменными с помощью оценки коэффициента корреляции и ковариации. аимосвязь двух или нескольких величин называется «корреляция». Как оценить численно, насколько тесна взаимосвязь? Для этого находят коэффициент корреляции Пирсона. Он помогает определить наличие линейной связи между величинами и принимает значения от -1 до 1. Если с ростом первой величины растёт вторая, то коэффициент корреляции Пирсона — положительный. Если при изменении одной величины другая остаётся прежней, то коэффициент равен 0. Если рост одной величины связан с уменьшением другой, коэффициент отрицательный. Чем ближе коэффициент корреляции Пирсона к крайним значениям 1 или -1, тем теснее линейная взаимосвязь. Если значение близко к нулю, значит, линейная связь слабая либо отсутствует вовсе. Бывает, что коэффициент нулевой не потому, что связи между значениями нет, а из-за того, что у неё более сложный, не линейный характер. Потому-то коэффициент корреляции такую связь не берёт. Коэффициент Пирсона находят методом corr(). Если смотреть по получившемуся датасету, в результате выполнения данной команды, самая тесная линейная взаимосвязь между heart_disease и old_peak, а самая слабая между index и fasting_bs.

```python
df.corr()
```

Подключение библиотек seaborn и numpy 
Построение тепловой карты корреляции для более качественной визуализации и анализа данных
С помощью данной тепловой карты мы можем проверить наши выводы, сделанные на основании данных из предыдущего датасета
Тепловая карта корреляции строится с помощью метода heatmap()

```python
import seaborn
import numpy as np
matrix = np.triu(df.corr())
seaborn.heatmap(df.corr(), annot = True, fmt='.1g', mask = matrix, vmin = -1, vmax = 1, center = 0)
```

Ковариация — это мера, которая количественно определяет силу и направление взаимосвязи между парой столбцов датафрейма.
В данном случае коварицация > 0 - это означает прямую связь (при росте одной величины растет и другая).

```python
x,y,n = df['age'],df['resting_bp'], df.shape[0]
mean_x, mean_y = sum(x) / n, sum(y) / n
cov_xy = (sum((x[k] - mean_x) * (y[k] - mean_y) for k in range(n))/ (n - 1))
print(f'Расчет ковариации: {cov_xy}')
```
