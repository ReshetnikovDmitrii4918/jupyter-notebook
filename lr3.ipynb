Необходимо выполнить кластеризацию методов иерархическим агломеративным методом. 
Видим, что всего образовалось 13 кластеров

3 основные характеристики: центр кластера - это среднее геометрическое место точек в пространстве переменных, радиус кластера - максимальное расстояние точек от центра кластера.
Моя гипотеза - 13 кластеров
Оптимальное количество кластеров - 5

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist
import numpy as np

df = pd.pivot_table(df, index=['age'], values = ['resting_bp'])

distance_mat = pdist(df) # pdist посчитает нам верхний треугольник матрицы попарных расстояний

Z = hierarchy.linkage(distance_mat, 'single') # linkage — реализация агломеративного алгоритма
plt.figure(figsize=(10, 5))
dn = hierarchy.dendrogram(Z, color_threshold=0.5)
```

Метод k-средних

Моя гипотеза - 10 кластеров

Видим, что $J(C_k)$ падает сильно при увеличении числа кластеров уже не так сильно – при изменении $k$ с 5 до 6. Значит, в данной задаче оптимально задать 5 кластеров.

Кластер имеет следующие математические характеристики: центр, радиус, размер кластера. 

```python
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(df)
    inertia.append(np.sqrt(kmeans.inertia_))

plt.plot(range(1, 10), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

В ходе выполнения данной лабораторной работы было произведено изучение алгоритмов и методов кластерного анализа на практике. А именно иерархический императивный метод и метод k-средних.
Метод k-средних, в отличие от задачи классификации или регресии, в случае кластеризации сложнее выбрать критерий, с помощью которого было бы просто представить задачу кластеризации как задачу оптимизации.
В случае метод распространен вот такой критерий – сумма квадратов расстояний от точек до центроидов кластеров, к которым они относятся. Необходимо, чтобы точки располагались кучно возле центров своих кластеров, но такое будет достигаться тогда, когда кластеров столько же, сколько и точек (то есть каждая точка – это кластер из одного элемента). Для решения выбора числа кластеров часто пользуются такой идеей: выбирают то число кластеров, начиная с которого описанный функционал падает не так быстро.
Иерархический императивный метод, По итогам выполнения такого алгоритма можно также построить замечательное дерево склеивания кластеров и глядя на него определить, на каком этапе нам было бы оптимальнее всего остановить алгоритм и тем самым вычислить оптимальное количество кластеров.
